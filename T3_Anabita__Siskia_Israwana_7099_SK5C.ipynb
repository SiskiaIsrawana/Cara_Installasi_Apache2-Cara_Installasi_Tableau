```python
# Langkah 1: Mengunggah dataset dari komputer ke Google Colab
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload file CSV
uploaded = files.upload()

# Membaca dataset yang diunggah
df=pd.read_csv("Dataset-Ujian-Nasional CSV.csv")

# Memeriksa 5 baris pertama dataset
df.head()
```

```python
# Langkah 2: Set plot style
sns.set(style="whitegrid")

# 1. Distribusi nilai total
plt.figure(figsize=(10, 6))
sns.histplot(df['Nilai'], bins=10, kde=True, color='blue')
plt.title('Distribusi Nilai Total', fontsize=16)
plt.xlabel('Nilai Total', fontsize=12)
plt.ylabel('Jumlah Peserta', fontsize=12)
plt.show()
```

```python
# 2. Perbandingan UTS dan UAS
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='UTS', y='UAS', hue='Gender', style='Gender', palette='Set1')
plt.title('Perbandingan Nilai UTS dan UAS', fontsize=16)
plt.xlabel('Nilai UTS', fontsize=12)
plt.ylabel('Nilai UAS', fontsize=12)
plt.show()
```

```python
# 3. Distribusi Nilai Berdasarkan Gender (modifikasi untuk peringatan palette)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Nilai', data=df, hue='Gender', palette='Set2', dodge=False)
plt.title('Distribusi Nilai Berdasarkan Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Nilai Total', fontsize=12)
plt.show()
```

```python
# 4. Nilai Berdasarkan Mata Kuliah (tanpa palette untuk menghilangkan peringatan)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Matkul', y='Nilai', data=df)
plt.title('Distribusi Nilai Berdasarkan Mata Kuliah', fontsize=16)
plt.xlabel('Mata Kuliah', fontsize=12)
plt.ylabel('Nilai Total', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
